<h1>STOCK RETURNS AND DASHBOARD<H1>

<h3>Question 1: Extract Tesla Stock Data

```python
import yfinance as yf
import pandas as pd
```

```python
tesla=yf.Ticker("TSLA")
tesla_data=tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

<h3>Question 2: Extract Tesla Revenue<h3>

```python
import requests
from bs4 import BeautifulSoup
```

```python
revenue=requests.get("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm").text
soup=BeautifulSoup(revenue,"html5lib")

tesla_revenue=pd.DataFrame(columns=["Date","Revenue"])

for row in soup.find("tbody").find_all("tr"):
    col=row.find_all("td")
    date=col[0].text
    revenue=col[1].text
    
    tesla_revenue=tesla_revenue.append({"Date":date, "Revenue":revenue}, ignore_index=True)
    
tesla_revenue["Revenue"]=tesla_revenue["Revenue"].str.replace(',|\$',"")

tesla_revenue.dropna(inplace=True)

tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]

tesla_revenue.tail()
```

<h3>Question 3: Extract GameStop Stock Data<h3>

```python
gamestop=yf.Ticker("GME")
gamestop_data=gamestop.history(period="max")
gamestop_data.reset_index(inplace=True)
gamestop_data.head()
```

<h3>Question 4: Extract GameStop Revenue<h3>

```python
revenue=requests.get("https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue").text
soup=BeautifulSoup(revenue,"html5lib")

gamestop_revenue=pd.DataFrame(columns=["Date","Revenue"])

for row in soup.find("tbody").find_all("tr"):
    col=row.find_all("td")
    date=col[0].text
    revenue=col[1].text
    
    gamestop_revenue=gamestop_revenue.append({"Date":date, "Revenue":revenue}, ignore_index=True)
    
gamestop_revenue["Revenue"]=gamestop_revenue["Revenue"].str.replace(',|\$',"")


gamestop_revenue.tail()
```

<h3>Question 5 and 6: Tesla and GameStop Stock and Revenue DashBoard<h3>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def make_graph(tesla_data, tesla_revenue, gamestop_data, gamestop_revenue):
    
    figure, axis = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
    figure.tight_layout()
    
    axis[0, 0].bar(tesla_data["Date"], tesla_data["Close"])
    axis[0, 0].set_title("Tesla Stock Data")
    
    axis[0, 1].bar(tesla_revenue.iloc[::-1]["Date"], tesla_revenue.iloc[::-1]["Revenue"])
    axis[0, 1].set_title("Tesla Revenue")
    
    axis[1, 0].bar(gamestop_data["Date"], gamestop_data["Close"])
    axis[1, 0].set_title("GameStop Stock Data")
    
    axis[1, 1].bar(gamestop_revenue.iloc[::-1]["Date"], gamestop_revenue.iloc[::-1]["Revenue"])
    axis[1, 1].set_title("GameStop Revenue")
    
    
make_graph(tesla_data, tesla_revenue, gamestop_data, gamestop_revenue)
```
